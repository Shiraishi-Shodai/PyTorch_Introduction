```python
import torch
```

## 3.7.1 ストレージのインデックス化

```python
points = torch.tensor([[4.0, 1.0], [5.0, 3.0], [2.0, 1.0]])
points.storage()
```

```python
points = torch.tensor([[4.0, 1.0], [5.0, 3.0], [2.0, 1.0]])
points_storage = points.storage()
points_storage[0] = 2.0
points
```

## 3.7.2 torchオブジェクトにしか存在しないメソッド
